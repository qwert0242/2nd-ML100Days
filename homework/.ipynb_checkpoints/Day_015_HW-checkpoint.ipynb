# numpy.random.random_sample(size=None)
在半開區間[0.0,1.0]中返回隨機浮點數。

```python
"""
Your Code Here
"""

matrix =2* np.random.random_sample((10,10))-1

plt.figure(figsize=(10,10))
"""
Your Code Here
"""

heatmap = sns.heatmap(matrix, xticklabels=1, yticklabels=False,cmap = plt.cm.RdYlBu_r, vmin = -1.0, annot = True, vmax = 1.0)

plt.show()
```


# uniform均勻分配


```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
# 把 NaN 數值刪去, 並限制資料上限為 1000 : 因為要畫點圖, 如果點太多，會畫很久!
# N_sample = 1000
# plot_data = plot_data.dropna().sample(n = N_sample)

matrix =  np.random.uniform(-1,1,[1000,3])  

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""


grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.distplot )
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```


# numpy.random.randn(d0, d1, ..., dn)常態分配
# numpy.random.normal(loc=0.0, scale=1.0, size=None)常態分配


```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = np.random.randn(1000,3)

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""

grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.distplot  )
grid.map_lower(sns.kdeplot, cmap = plt.cm.RdYlBu_r)

plt.show()
```
