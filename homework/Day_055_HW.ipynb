- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# Though the following import is not directly being used, it is required
# for 3D projection to work
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
%matplotlib inline

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

## 流程
## 前提：在知道為k個類別的情況下

### 隨機在數據當中抽取k個樣本當作k個類別的中心點
### 計算其餘的點到k個中心點的距離
### 其餘的每一個點都會存著k個與中心點之間的距離
### 從中選出距離最近的一個點作為自己類別的標記，形成k個族群
### 分別計算這k個族群的平均值
### 把k個族群的平均值點 與 k個舊中心點進行比較
### 比較結果相同：兩點重合，結束聚類
### 比較結果不同：把這k個平均值點當作新的中心點，從第2步開始執行

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

## 設定 模型 估計參數

```python
estimators =[('k_means_8',KMeans(n_clusters=8)),
             ('k_means_3',KMeans(n_clusters=3)),
             ('k_means_init',KMeans(n_clusters=3,n_init=5,init='random'))]

#n_init 设为10意味着进行10次随机初始化，选择效果最好的一种来作为模型
#init=’k-means++’ 会由程序自动寻找合适的n_clusters； 
```

## 資料建模 並 視覺化 結果

```python
fignum = 1 #建立圖的編號
titles = ['8 clusters','3 clusters','3 clusters,5 initialization']
for name, est in estimators:
    fig = plt.figure(fignum,figsize=(5,4))
    ax = Axes3D(fig, rect=[0,0, .95,1], elev=48, azim=134)
    
    ##fit data
    est.fit(X)
    labels = est.labels_
    ax.scatter(X[:, 3],X[:, 0],X[:, 2],   #選三個特徵出來看
              c=labels.astype(np.float),edgecolor='k')
    # 去除scale
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_title(titles[fignum - 1]) #減1 是要取list裡面的字串(index從0開始)
    ax.dist = 12
    fignum = fignum + 1
#Axes3D : 定義一個3D的圖形 est.fit : 根據上面 estimators 去 fit 資料庫的圖 ax.scatter : 畫散點圖，
#後面的參數用來調整顏色 ax.dist : 設定與物體之間的距離
    
```

```python
# Plot the ground truth
fig = plt.figure(fignum, figsize=(4, 3))
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

for name, label in [('cls0', 0),
                    ('cls1', 1),
                    ('cls2', 2)]:
    ax.text3D(X[y == label, 3].mean(),
              X[y == label, 0].mean(),
              X[y == label, 2].mean() + 2, name,   #show字在圖上的座標移位
              horizontalalignment='center',
              bbox=dict(alpha=.2, edgecolor='w', facecolor='w'))
#np.choose : 將原本 label 順序的(0 1 2)改成(1 2 0) ax.text3D : 將不同label的資料標上個別物種類名稱，
#裡面 X[y == label, 3].mean() 用在調整 text 的 X Y Z 軸位置    
# Reorder the labels to have colors matching the cluster results
y = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X[:, 3], X[:, 0], X[:, 2], c=y, edgecolor='k')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_title('Ground Truth')
ax.dist = 12

#fig.show()
```
