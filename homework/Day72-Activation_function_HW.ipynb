# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
x =  np.arange(-10,10,0.01)
y=list(map(lambda x:x if x>0 else 0,x))

#ReLU函数图像
plt.plot(x,y)
plt.grid(color='b' , linewidth='0.1' ,linestyle='--')

#ReLU函数导数图像
dev_y =list(map(lambda x: 1 if x>0 else 0,x))
plt.plot(x,dev_y)
plt.grid(color='b' , linewidth='0.1' ,linestyle='--')
```

```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

#plt.gca 获取当前坐标轴信息
#plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))设置刻度线

#relu 數學函數表示方式
#relu= (lambda x:x if x>0 else 0,x)
#drelu=(lambda x: 1 if x>0 else 0,x)
def relu(x):
    return np.maximum(x,0)

#relu 微分
def drelu(x):
    return (1 * (x > 0))

plt.figure(figsize=(8,8))

# linespace generate an array from start and stop value
# with requested number of elements. Example 10 elements or 100 elements.
x = plt.linspace(-10,10,100)

plt.subplot(2,1,1)   

# prepare the plot, associate the color r(ed) or b(lue) and the label 
plt.plot(x, relu(x), 'b', label='linspace(-10,10,10)')

# Draw the grid line in background.
plt.grid()

# 顯現圖示的Title
plt.title('Relu')

# 顯現 the relu formula
plt.text(4, 0.8, r'f(x)=max(0,x)', fontsize=15)

#resize the X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))
 
    
plt.subplot(2,1,2)

# prepare the plot, associate the color r(ed) or b(lue) and the label 
plt.plot(x, drelu(x), 'b', label='linspace(-10,10,10)')

# Draw the grid line in background.
plt.grid()

# 顯現圖示的Title
plt.title('dReLU')


#resize the X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.1))

# create the graph
plt.show()
```
